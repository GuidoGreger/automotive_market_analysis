###  Dataset Description
You are given a car dataset with the following columns:
- **Make:** The manufacturer of the car (e.g., Toyota, Ford, BMW).
- **Model:** The specific model name of the car.
- **Year:** The production year of the car.
- **Engine Fuel Type:** Type of fuel the car uses (e.g., gas, diesel, electric, hybrid).
- **Engine HP:** The horsepower of the car's engine.
- **Engine Cylinders:** The number of cylinders in the car's engine.
- **Transmission Type:** The type of transmission in the car (e.g., automatic, manual).
- **Driven_Wheels:** The type of drivetrain (e.g., front-wheel drive, rear-wheel drive, all-wheel drive).
- **Number of Doors:** Number of doors the car has (e.g., 2, 4).
- **Market Category:** The market segment the car belongs to (e.g., SUV, sedan, luxury).
- **Vehicle Size:** The size category of the car (e.g., compact, mid-size, large).
- **Vehicle Style:** The style of the vehicle (e.g., coupe, convertible, hatchback).
- **highway MPG:** The estimated miles per gallon on the highway.
- **city mpg:** The estimated miles per gallon in the city.
- **Popularity:** A numerical value indicating how popular the car is in the market.
- **MSRP:** The Manufacturer's Suggested Retail Price of the car.

Your goal is to **clean, transform, and analyze the data** by performing the tasks below.

#### Analyzing MSRP

```python
# Creating a categorical variable for Engine HP along the quartiles
df_cars.loc[:,'MSRP Category'] = pd.cut(df_cars['MSRP'], bins=[0, 23120, 31630, 44000, 2065902], labels=["Min - 25%", "25% - 50%", "50% - 75%", "75% - Max"])

# Boxplot to visualize the distribution of values across Engine HP categories
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_cars, x='MSRP Category', y='MSRP', hue='MSRP Category', palette='viridis')

plt.xlabel("MSRP Percentiles")
plt.ylabel("MSRP Value")
plt.suptitle("Boxplot of MSRP by Percentile Categories", fontsize=14)
plt.show()
```

```python
# Creating seperate boxplots for bottom 3 and top percentiles and show topmost data

import matplotlib.ticker as mticker

# Define bins for bottom 3 percentiles (excluding extreme values)
bins_main = [0, 23120, 31630, 44000]
labels_main = ['Min - 25%', '25% - 50%', '50% - 75%']

# Create a categorical variable for the main range
df_cars['MSRP Percentile'] = pd.cut(df_cars['MSRP'], bins=bins_main, labels=labels_main).astype('category')

# Separate the 4th bin (MSRP > 44,000)
df_top_percentile = df_cars[df_cars['MSRP'] > 44000].copy()
df_top_percentile['MSRP Percentile'] = '75% - Max'

# Create the figure and axes
fig, ax = plt.subplots(1, 2, figsize=(12, 6), sharey=False)

# Boxplot for the main MSRP categories
sns.boxplot(data=df_cars, x='MSRP Percentile', y='MSRP', hue='MSRP Percentile', palette='viridis', ax=ax[0])
ax[0].set_title('MSRP Distribution (Bottom 3 Percentiles)')

# Boxplot for the extreme values separately
sns.boxplot(data=df_top_percentile, x='MSRP Percentile', y='MSRP', hue='MSRP Percentile', palette='magma', ax=ax[1])
ax[1].set_title('MSRP Distribution (Top Percentile)')

# Function to format y-axis labels as full numbers
def ylabel_format(x, pos):
    return f"{x:,.0f}"

# Apply the formatter to both subplots
ax[0].yaxis.set_major_formatter(mticker.FuncFormatter(ylabel_format))
ax[1].yaxis.set_major_formatter(mticker.FuncFormatter(ylabel_format))

plt.tight_layout()
plt.show()
```

The opposite is true for fuel-driven cars that use less fuel when driving steadily on a highway as opposed to frequent accelaration and decelaration in the city which consumes more fuel.

### 4. Correlation Analysis

Investigate the correlation between the following variables:

Engine HP, MSRP, Popularity, city mpg, and highway MPG.

```python
correlation_table = df_cars_filtered[['Engine HP','MSRP','Popularity','city mpg','highway MPG']].corr(numeric_only=True)

plt.figure(figsize=(11, 7))
sns.heatmap(correlation_table, annot=True, cmap="RdBu", center=0, vmin=-1, vmax=1, linewidths=0.5, square=True)

plt.title('Correlation Heatmap of selected metrics')
plt.show()
```

Correlations between the 5 metrics range between **-0.0018** (city mpg / Popularity) and **0.89** (highway MPG / city mpg):

- Very strong correlations (0.90 to 1.00 // -0.90 to -1.00)
  - highway MPG / city mpg: 0.94

- Moderate correlations (0.40 to 0.69 // -0.40 to -0.69)
  - MSRP / Engine HP: 0.65

- Weak correlations (0.10 to 0.39 // -0.10 to -0.39)
  - city mpg / Engine HP: -0.37
  - highway mpg / Engine HP: -0.40
  - highway mpg / MSRP: -0.19
  - city mpg / MSRP: -0.17

- No correlations (0.00 to 0.09 // -0.00 to -0.09)
  - Popularity / MSRP: -0.058
  - Popularity / Engine HP: 0.03
  - highway MPG / Popularity: -0.024
  - city mpg / Popularity: -0.0017

The correlation heatmap confirms some obious assumptions:
- an engine that uses more fuel than others will do so no matter if it operates in a city or on a highway
- cars with stronger enginges tend to be more expensive than cars with weaker engines
- Engine HP is only weakly correlated to city or highway mpg, but in an expected manner: more HP goes along with lower mpg
- The is valid for MSRP and city of highway mpg, suggesting that more expensive cars tend achieve less mileage (possibly due to stronger, bigger engines in more expensive cars)
- Popularity does not correlate with any of the other metrics. The popularity of certain technical values depend more on the customer segment and their attitudes.
